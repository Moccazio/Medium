# Ken French's Database

```python
import numpy as np
import pandas as pd

import datetime as dt
from datetime import datetime
from datetime import datetime, timedelta 

date = dt.datetime.now()
modified_date = date - timedelta(days=1)
end = datetime.strftime(modified_date, "%Y/%m/%d")

import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = (16.0, 8.0)

#French Database
import pandas_datareader.data as web  
from pandas_datareader.famafrench import get_available_datasets
```

```python
datasets = get_available_datasets()
print('No. of datasets:{0}'.format(len(datasets)))
```

## Industry Portfolios

## Fama and French 12 Industry Classification

Industry
1. NoDur = Consumer NonDurables -- Food, Tobacco, Textiles, Apparel, Leather, Toys
2. Durbl = Consumer Durables -- Cars, TV's, Furniture, Household Appliances
3. Manuf = Manufacturing -- Machinery, Trucks, Planes, Off Furn, Paper, Com Printing
4. Enrgy = Energy -- Oil, Gas, and Coal Extraction and Products
5. Chems = Chemicals -- Chemicals and Allied Products
6. BusEq = Business Equipment -- Computers, Software, and Electronic Equipment
7. Telcm = Telecommunication -- Telephone and Television Transmission
8. Utils = Utilities
9. Shops = Wholesale, Retail, Laundries and Repair Shops
10. Hlth = Health -- Healthcare, Medical Equipment, and Drugs
11. Money = Finance
12. Other = Mines, Constr, BldMt, Trans, Hotels, Bus Serv, Entertainment

```python
df_12_industry = [dataset for dataset in datasets if '12' in dataset and 'Industry' in dataset]
print(df_12_industry)

```

```python
ds_industry = web.DataReader(df_12_industry[0],'famafrench',start='1963-07-01',end=end) 
```

# Risk factor dataset

We perform the same process above Fama-French risk factor portfolio dataset

```python
df_5_factor = [dataset for dataset in datasets if '5' in dataset and 'Factor' in dataset]
print(df_5_factor)
ds_factors = web.DataReader(df_5_factor[0],'famafrench',start='1963-07-01',end=end) # Taking [0] as extracting 1F-F-Research_Data_Factors_2x3')
print('\nKEYS\n{0}'.format(ds_factors.keys()))
print('DATASET DESCRIPTION \n {0}'.format(ds_factors['DESCR']))
ds_factors[0].head()
```
